```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('query.csv')
df.info()
```

```python

df.head()
```

Заполняем пропуски нулями

```python
df.fillna(0, inplace=True)
df.head()
```

```python
df.describe()
```

Затраты на рейс считаем по следующей формуле:


(средний расход топлива * длительность полета * стоимость топлива/1000) * 100/23


*стоимость топлива в исходных данных приведена за 1 тонну, нам нужно за килограмм

*считаем, что затраты на топливо составляют 23% от всех затрат на рейсе

```python
df['cost'] = round((df.consump_fuel*df.duration*df.cost_fuel/1000)*100/23)
df.head()
```

Считаем прибыль рейса, вычитая из суммарного дохода общие затраты на рейс

```python
df['profit'] = df.revenue - df.cost
df.head()
```

```python
df.profit = df.profit/1000
```

```python
df.head()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
sns.barplot(x="month", y="profit", hue="flight_no",
            data=df, order=[12, 1, 2])
ax.set_xlabel('Декабрь, Январь, Февраль')
ax.set_ylabel('Прибыль, тыс. руб')
ax.set_title('Средняя прибыль по рейсам по месяцам')
```

Найдем среди рейсов, приносящих прибыль, 5% самых малоприбыльных

```python
flight_0252 = df[df.flight_no == 'PG0252'][['flight_id', 'flight_no',
                                                    'month', 'duration',
                                                    'tickets_sold',
                                                    'max_seats', 'revenue','cost','profit']].copy()
```

```python
quantile = flight_0252.profit.quantile(q=0.05)
flight_0480_kill = flight_0252[flight_0252.profit < quantile].sort_values(
    by='profit', ascending=True)
flight_0480_kill
```
